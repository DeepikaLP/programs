```python
#Simple calculator
Num1 =float(input("enter the value"))
operator=input("Enter the operator (+,-,*,/)")
Num2=float(input("Enter the value"))
if operator =='+':
  print(Num1 + Num2)
elif operator =='-':
  print(Num1 - Num2)
elif operator =='*':
  print(Num1 * Num2)
elif operator =='/':
  print(Num1 /Num2)
else:
  print("Invalid")
```

```python
#The program picks a random number between 1â€“100, and the user has to guess it

secret_number=random.randint(1,10)
while True:
  guess=int(input("Enter a number"))
  if guess < secret_number:
    print("Too small")
  elif guess > secret_number:
    print("Too big")
  else:
    print("Guessed correct")
    break
```
